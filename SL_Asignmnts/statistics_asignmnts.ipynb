2. Find the mean of first 10 Fibonacci numbers (Use a for loop to create 10 Fibonacci series)

```python
a=0
b=1
print(a,b,end=" ")
for i in range(2,10):
    k=a+b
    a=b
    b=k
    print(k,end = " ")
```

3. Find the mean and median of first 5 prime numbers.

```python
lw = 1
up = 12
for num in range(lw, up + 1):
   if num > 1:
       for i in range(2, num):
           if (num % i) == 0:
               break
       else:
           print(num, end=" ")
```
